```python
df_forCompareUp = df_forCompareUp.assign(tempName = lambda df:df['Cluster'] + ', ' + df['Time'] + '|' + df['Project'])
```

```python
ls_allComb = df_forCompareUp['tempName'].unique()
```

```python
from itertools import product
_iter = list(product(ls_allComb, ls_allComb))
df_shareCounts = pd.DataFrame(index=ls_allComb, columns=ls_allComb)
df_shareRatio = pd.DataFrame(index=ls_allComb, columns=ls_allComb)
for x,y in _iter:
    if x == y:
        continue
    shareCounts = len(set(df_forCompareUp.query('tempName == @x').dropna().mt4) & set(df_forCompareUp.query('tempName == @y').dropna().mt4))
    shareRatio = shareCounts / len(df_forCompareUp.query('tempName == @x').dropna().mt4)
    df_shareCounts.loc[x,y] = shareCounts
    df_shareRatio.loc[x,y] = shareRatio
```

```python
fig, ax = plt.subplots(figsize=(14,13))
sns.heatmap(df_shareRatio.fillna(0), cmap='Reds', annot=True, fmt='.1%', mask=np.eye(len(df_shareRatio)))
plt.xticks(ax.get_xticks(), [x.get_text().split('|')[0] for x in ax.get_xticklabels()], fontsize=14)
plt.yticks(ax.get_yticks(), [x.get_text().split('|')[0] for x in ax.get_yticklabels()], fontsize=14, rotation=0)

# plt.xticks(ax.get_xticks(), [x for x in ax.get_xticklabels()], fontsize=14)
# plt.yticks(ax.get_yticks(), [x for x in ax.get_yticklabels()], fontsize=14, rotation=0)

plt.show()
```
